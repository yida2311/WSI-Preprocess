# 5x->2.5x for WSI

```python
import os
from PIL import Image
from tqdm import tqdm_notebook as tqdm

root = '/media/ldy/7E1CA94545711AE6/OSCC/'

src = '5x_mask_fine/std_mask'
dst = '2.5x_mask_fine/std_mask'

src_dir = os.path.join(root, src)
dst_dir = os.path.join(root, dst)
if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)

slide_list = sorted(os.listdir(src_dir))
# slide_list = ['2020-07912.png', '2020-07912-2.png', '2017-05588.png']
for c in tqdm(slide_list):
    src_path = os.path.join(src_dir, c)
    dst_path = os.path.join(dst_dir, c)
#     print(src_dir)
    src_img = Image.open(src_path)
    w, h = src_img.size
    dst_img = src_img.resize((w//2, h//2), Image.NEREST)
    dst_img.save(dst_path)
```

# 5x -> 2.5x for Tile

```python
import os
import json
from tile import Tiler

file_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_filter/filtered_img'
std_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_fine_mask/std_mask'
rgb_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_fine_mask/rgb_mask'
sample_type = 'seg'
```

```python
# slide_list = cnt['train']
slide_list = sorted([c.split('.')[0] for c in os.listdir(std_mask_dir)])
# slide_list = ['2020-07912', '2020-07912-2', '2017-05588']
save_patch_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_tile/temp/patch'
save_std_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_tile/temp/std_mask'
save_rgb_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC_FINE/2.5x_tile/temp/rgb_mask'
tiler = Tiler(slide_list,
              file_dir,
              std_mask_dir=std_mask_dir,
              rgb_mask_dir=rgb_mask_dir,
              save_patch_dir=save_patch_dir,
              save_std_mask_dir=save_std_mask_dir,
              save_rgb_mask_dir=save_rgb_mask_dir,
              sample_type=sample_type,
              scale_factor=1,
             )

patch_size = 640
overlap = 80
filter_rate = 0.02

tiler.sample_and_store_patches_png(patch_size,
                                   overlap,
                                   filter_rate)
```
